This notebook takes the clean_relevant_db.csv and leaves it only with the top 10% values, ranked by total relevance (the sum of both relevance columns).
This is performed in order to have a more reasonable sized Gephi Graph.

```python
import pandas as pd
```

```python
file_path = '../03_clean_dbs/clean_relevant_db.csv'
clean_df = pd.read_csv(file_path, index_col=0)
clean_df['total_relevance'] = clean_df['ukraine_relevance'] + clean_df['russia_relevance']
ten_percent_of_rows = clean_df.shape[0] // 10 # Get the number of rows divided by 10
most_relevant_values_df = clean_df.nlargest(ten_percent_of_rows, 'total_relevance')
most_relevant_values_df
```

```python
csv_path = 'most_relevant_values.csv'
most_relevant_values_df.to_csv(csv_path, encoding='utf-8')
```
